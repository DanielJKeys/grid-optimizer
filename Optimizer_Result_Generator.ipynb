```python
df_list = []
for f in file_list:
  df_list.append(pd.read_csv(f"/content/drive/Shareddrives/Autonomy Prime (IL4)/5. Tech Transition Team/DevOps/Optimizer Model/Model_Save_X{x}_Y{y}_Z{z}/" + f))
```

#Function to remove duplicates from list

```python
def remove_Duplicates(k):
    newk = []
    for i in k:
        if i not in newk:
            newk.append(i)
    return newk
```

#Find Unique Radii

```python
unique_radii = []
for df in df_list:
  unique_radii.extend(df["Radius"].unique())
unique_radii = remove_Duplicates(unique_radii)
unique_radii = sorted(unique_radii)
```

#Pull Metrics

```python
for df in df_list:
  metric_list.append(sum(df["Volume"]))
  cone_list.append(df[df["Shape"] == 'Cone'].shape[0])
  sphere_list.append(df[df["Shape"] == 'Sphere'].shape[0])
  red_list.append(df[df["Type"] == 'R'].shape[0])
  blue_list.append(df[df["Type"] == 'B'].shape[0])
  temp_radii = []
  for radius in unique_radii:
    temp_radii.append(len(df[df["Radius"] == radius]))

    radius_list.append(temp_radii)
```

#Health Check

```python
print(f"Percentage Red: {sum(red_list) / (sum(red_list) + sum(blue_list))}")
print(f"Percentage Cones: {sum(cone_list) / (sum(sphere_list) + sum(cone_list))}")
```

#Save Results

```python
with open(f"/content/drive/Shareddrives/Agility Prime Hub (IL4)/Program Management/Configuration_and_Data_Management/03_Optimizer_Model/Results/Results_X{x}_Y{y}_Z{z}.csv", 'w+') as f:
  f.write(f"Model Number,Cost,Spheres,Cones,Red,Blue,{','.join([str(radius) for radius in unique_radii])}\n")
  for i in range(len(metric_list)):
    f.write(f"{i},{metric_list[i]},{sphere_list[i]},{cone_list[i]},{red_list[i]},{blue_list[i]}, {','.join([str(radius) for radius in radius_list[i]])}\n")
```
