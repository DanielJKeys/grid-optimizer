# Red vs Blue:

```python
rvb_dict = {"Blue":sum(df["Blue"]), "Red":sum(df["Red"])}
px.pie(title="Red vs Blue", values=[v for v in rvb_dict.values()], names=[l for l in rvb_dict.keys()], color=[l for l in rvb_dict.keys()], color_discrete_map={'Red':'#d62728', 'Blue':'#1f77b4'})
```

# Spheres vs Cones

```python
svc_dict = {"Sphere":sum(df["Spheres"]), "Cone":sum(df["Cones"])}
px.pie(title="Sphere vs Cones", values=[v for v in svc_dict.values()], names=[l for l in svc_dict.keys()], color=[l for l in svc_dict.keys()], color_discrete_map={'Sphere':'#7f7f7f', 'Cone':'#9467bd'})
```
